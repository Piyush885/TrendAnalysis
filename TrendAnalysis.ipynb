```python
df2 = pd.DataFrame({'sentences': a})
```

```python
def tokenize_and_lemmatize(text):
    tokens = word_tokenize(text)  # Tokenize sentence into words
    lemmatizer = WordNetLemmatizer()
    lemmas = [lemmatizer.lemmatize(token) for token in tokens]  # Lemmatize tokens
    return lemmas
```

```python
df['lemmas'] = df['sentences'].apply(tokenize_and_lemmatize)
```
